## Based on the analysis of ratings and votes, which genres of movies have demonstrated the highest performance? Considering this performance, which specific genre of movie should Microsoft prioritize for production?
In this analysis, I will thoroughly analyze the data to identify the most successful genres. As the genre is a broad category, this exploration will provide us with a diverse range of movie options to consider producing once we identify the genres that have been particularly prosperous in the movie industry.

```python
# Merge the DataFrames based on the common column 'tconst'
merged_data = pd.merge(title_basics, title_rating, on='tconst')

# Select specific columns that are required
top_rated = merged_data[['original_title', 'genres', 'averagerating', 'numvotes']]

# Sort the DataFrame in descending order based on 'averagerating'
top_rated = top_rated.sort_values(by='averagerating', ascending=False)

top_rated.head(20)
```

### SELECTING THE TOP GENRES

```python
# Extract the genres column
genres_column = top_rated['genres']

# Split the genres string and create a list of all genres
all_genres = []
for genres in genres_column:
    all_genres.extend(genres.split(','))

# Count the occurrences of each genre
genre_counts = pd.Series(all_genres).value_counts()

# Sort the genres based on their occurrences in descending order
sorted_genres = genre_counts.sort_values(ascending=False)

# Select the top ten genres
top_genres = sorted_genres.head(10)

top_genres
```

### PLOTTING A GRAPH OF THE TOP GENRES
The pie chart is used to visualize the distribution of the top genres and the proportion of each genre relative to the total number of movies. Each slice of the pie represents a genre, and its size corresponds to the percentage of movies that belong to that genre. 

```python
# Plotting the top genres using a pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_genres, labels=top_genres.index, autopct='%1.1f%%', startangle=140, shadow=True)
plt.axis('equal')
plt.title('Top Genres Distribution')
plt.show()
```

From the pie chart, we can observe the distribution of the top genres based on the number of movies. Drama stands out as the most prevalent genre, making up the largest slice of the pie. It is followed by comedy, action, thriller, adventure, horror, and other genres.

Using a function, get the top genres in the foreign and domestic market

```python
# Define the function to get the top genres in the foreign and domestic market
def get_top_genres(df, column_name, top_ten=10):
    genres_column = df[column_name]
    all_genres = [genre for genres in genres_column for genre in genres.split(',')]
    genre_counts = pd.Series(all_genres).value_counts()
    sorted_genres = genre_counts.sort_values(ascending=False)
    top_genres = sorted_genres.head(top_ten)
    return top_genres
```

```python
# Call the function to get the top genres for the domestic market
top_genres_domestic = get_top_genres(dom_rating_gross_data, 'genres')
print("Top Genres (Domestic):",top_genres_domestic)
```

```python
# Call the function to get the top genres for the domestic and foreign dataframes
top_genres_foreign = get_top_genres(for_rating_gross_data, 'genres')
print("Top Genres (Foreign):",top_genres_foreign)
```

```python
# Create figure and axes
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 8))

# Plot domestic genres with pastel colors on ax1
ax1.bar(top_genres_domestic.index, top_genres_domestic.values, color='skyblue')
ax1.set_title('Top Genres in Domestic Market')
ax1.set_xlabel('Genres')
ax1.set_ylabel('Number of Movies')

# Plot foreign genres with lightgreen color on ax2
ax2.bar(top_genres_foreign.index, top_genres_foreign.values, color='lightgreen', alpha=0.7)
ax2.set_title('Top Genres in Foreign Market')
ax2.set_xlabel('Genres')
ax2.set_ylabel('Number of Movies')

# Rotate x-axis labels for better readability
ax1.tick_params(axis='x', rotation=45)
ax2.tick_params(axis='x', rotation=45)

# Adjust the layout to avoid overlapping titles and labels
plt.tight_layout()

plt.show()
```

Based on the graph, the top genres in the domestic market are Drama, Comedy, Action, Adventure, Thriller, Romance, Crime, Horror, Biography, and Mystery. These genres have varying numbers of movies released in the domestic market, with Drama being the most prevalent with over 700 movies.
In the foreign market, the top genres are Drama, Comedy, Action, Adventure, Romance, Thriller, Crime, Biography, Horror, and Mystery. Similarly, Drama is the most dominant genre in the foreign market with over 900 movies.

Both the domestic and foreign markets share similar top genres, with Drama, Comedy, and Action being the top three in both cases.

# Data Cleaning

```python
# Remove dollar signs and commas from each column
recent_movie_budget = recent_movie_budget.applymap(lambda x: x.replace('$', '').replace(',', '') if isinstance(x, str) else x)
recent_movie_budget.head()
```

```python
# Change the data type of the production_budget, domestic_gross, worldwide_gross and movie
recent_movie_budget['production_budget'] = pd.to_numeric(recent_movie_budget['production_budget'])
recent_movie_budget['domestic_gross'] = pd.to_numeric(recent_movie_budget['domestic_gross'])
recent_movie_budget['worldwide_gross'] = pd.to_numeric(recent_movie_budget['worldwide_gross'])
```

```python
recent_movie_budget.info()
```

Creating profit column

```python
#Create the profit columns 
recent_movie_budget['Profit'] = (recent_movie_budget['worldwide_gross'] - recent_movie_budget['production_budget'])
recent_movie_budget
```

```python
# Remove all losses from the dataframe(profits is negative)
recent_movie_budget = recent_movie_budget[recent_movie_budget['Profit'] > 0]
recent_movie_budget.head()
```

```python
# correlation between production_budget, domestic_gross, worldwide_gross and Profit
# Calculate the correlation matrix
correlation_matrix = recent_movie_budget[['production_budget', 'domestic_gross', 'worldwide_gross', 'Profit']].corr()

# Display the correlation matrix
correlation_matrix
```

The results from the correlation matrix suggest that there is a positive relationship between the production budget of a movie and its domestic and worldwide gross revenues, as well as its overall profit.

```python
# However we are trying to figure out the optimal budget in order to realize the highest profit
# Create a scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=recent_movie_budget, x='production_budget', y='Profit', color='blue', marker='o')

# Format y-axis labels in dollars
plt.gca().yaxis.set_major_formatter('${:,.0f}'.format)
plt.gca().xaxis.set_major_formatter('${:,.0f}'.format)

# Format x-axis labels
plt.xticks(rotation=60)

# Add labels and title
plt.xlabel('Production Budget')
plt.ylabel('Profit')
plt.title('Profit vs. Production Budget for Recent Movies')

plt.tight_layout()
plt.show()
```

From the graph there are a significant number of movies with low production budgets and low profits. This suggests that there are many movies that were produced with relatively small budgets and did not generate substantial profits.
